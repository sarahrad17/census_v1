```python

#import packages
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from collections import Counter

#counties = ['Alameda', 'Armstrong', 'Cibola', 'Fayette', 'GrandForks', 'Hawaii', 'Jefferson', 'Nantucket', 'Washington']
counties = ['Alameda', 'GrandForks', 'Armstrong']
plt.figure(figsize=(15,8))
for county in counties:
    if county == 'Alameda':
        county1 = 'Hawaii'
        county2 = 'Cibola'
        legendval = 'High Diversity Population'
    if county == 'GrandForks':
        county1 = 'Fayette'
        county2 = 'Nantucket'
        legendval = 'Medium Diversity Population'
    if county == 'Armstrong':
        county1 = 'Jefferson'
        county2 = 'Washington'
        legendval = 'Low Diversity Population'
    #dp_df = pd.read_csv (r'data/'+county+'/dp/dp.csv')
    swap_x = []
    swap_y = []
    swap_y1 = []
    swap_y2 = []

    #swap_df = pd.read_csv (r'accuracy_data/'+county+'/random/random.csv')
    swap_df = pd.read_csv (r'mean_dif/'+county+'/dp/dp.csv')
    #for i, row in dp_df.iterrows():
    nn=0
    for i, row in swap_df.iterrows():
        swap_x.append(float(row['filename']))
        swap_y.append(row['mean_difference'])

    swap_df = pd.read_csv (r'mean_dif/'+county1+'/dp/dp.csv')
    for i, row in swap_df.iterrows():
        swap_y1.append(row['mean_difference'])

    
    swap_df = pd.read_csv (r'mean_dif/'+county2+'/dp/dp.csv')
    for i, row in swap_df.iterrows():
        swap_y2.append(row['mean_difference'])
    swapy= []           
    for v in range(0,len(swap_x)):
        swapy.append((swap_y[v]+swap_y1[v]+swap_y2[v])/3)
        

    plt.rcParams.update({'font.size': 18})
    
    axes= plt.axes()
    plt.xlabel('Epsilon Value')
    plt.ylabel('Mean Difference From True Value')
    plt.title('Differential Privacy: Epsilon Value and Mean Difference For Racial Minority Representation')
    plt.plot([i * 5 for i in swap_x], swapy, label=legendval)
    handles, labels = axes.get_legend_handles_labels()
    axes.legend(handles, labels, loc='upper right')
    plt.savefig('test1.png')
```
