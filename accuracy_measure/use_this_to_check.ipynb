```python
import os
from collections import Counter

counties = ['Alameda', 'Armstrong', 'Cibola', 'Fayette']
plt.figure(figsize=(15,8))
for county in counties:
    #dp_df = pd.read_csv (r'data/'+county+'/dp/dp.csv')
    swap_x = []
    swap_y = []


    #swap_df = pd.read_csv (r'accuracy_data/'+county+'/random/random.csv')
    dp_df = pd.read_csv (r'mean_dif/'+county+'/dp/dpnewnew.csv')
    #for i, row in dp_df.iterrows():
    nn=0
    for i, row in dp_df.iterrows():
        swap_x.append(float(row['filename']))
        swap_y.append(row['mean_difference'])

        

    plt.rcParams.update({'font.size': 18})
    
    axes= plt.axes()
    plt.xlabel('Epsilon Value')
    plt.ylabel('Mean Difference From True Value')
    plt.title('Differential Privacy: Epsilon Value and Mean Difference For Racial Minority Representation')
    plt.plot([i * 5 for i in swap_x], swap_y, label=county)
    handles, labels = axes.get_legend_handles_labels()
    axes.legend(handles, labels, loc='upper right')
    plt.savefig('test1.png')
```
