```python
##########################################################################################################################
#CREATE REAL ORDER
##########################################################################################################################
def create_real_order(state):
    #read in race data
    df_r = pd.read_csv (r'../swapping/raw_state_data/'+state+'/race_data.csv')
    #read in hispanic data
    df_h = pd.read_csv (r'../swapping/raw_state_data/'+state+'/hispanic_data.csv').T
    
    race_total = []
    total=0
    c = 0
    for (columnName, columnData) in df_r.iteritems():
        if c > 0:
            val = float(str(columnData.values).replace('\'','').replace('[','').replace(']','').replace(',',''))
            total = total+val
        c+=1
    c=0
    for (columnName, columnData) in df_r.iteritems():
        if c>0:
            val = float(str(columnData.values).replace('\'','').replace('[','').replace(']','').replace(',',''))
            race_total.append([columnName.replace('\xa0',''), val/total * 100])
        c+=1
        
    #hisp total will be an array of race types' hispanic proportions.
    #ex. hisp_total[0] is for white alone; [not hispanic, hispanic] ex. [.1, .9]
    hisp_total = [[],[],[],[],[],[],[], []]
    for row in df_h.iterrows():
            if 'White alone' in row[0]:
                hisp_total[0].append(row[1].iloc[0])
            elif 'Black or African American alone' in row[0]:
                hisp_total[1].append(row[1].iloc[0])
            elif 'American Indian and Alaska Native alone' in row[0]:
                hisp_total[2].append(row[1].iloc[0])
            elif 'Asian alone' in row[0]:
                hisp_total[3].append(row[1].iloc[0])
            elif 'Native Hawaiian and Other Pacific Islander alone' in row[0]:
                hisp_total[4].append(row[1].iloc[0])
            elif 'Some Other Race alone' in row[0]:
                hisp_total[5].append(row[1].iloc[0])
            elif 'Two races including Some other race' in row[0]:
                hisp_total[6].append(row[1].iloc[0])
            elif 'Two races excluding Some other race, and three or more races' in row[0]:
                hisp_total[7].append(row[1].iloc[0])

        
    #final race will be an array of race types' hispanic proportions.
    #ex. finalrace[0] is for white alone; [not hispanic, hispanic] ex. [.1, .9]
    
    finalrace = [ [] for _ in range(126) ]
    
    curr = 0
    for rval in race_total:
        r = rval[0]
        if 'White alone' in r:
            hval = hisp_total[0]
        elif 'Black or African American alone' in r:
            hval = hisp_total[1]
        elif 'American Indian and Alaska Native alone' in r:
            hval = hisp_total[2]
        elif 'Asian alone' in r:
            hval = hisp_total[3]
        elif 'Native Hawaiian and Other Pacific Islander alone' in r:
            hval = hisp_total[4]
        elif 'Some other race' in r and 'Two or More Races' in r:
            hval = hisp_total[6] 
        elif 'Some Other Race alone' in r:
            hval = hisp_total[5]
        elif 'Two or More Races' in r:
            hval = hisp_total[7]
        r1 = rval[1] * hval[0]
        r2 = rval[1] * hval[1]
        finalrace[curr+63] = ['Not Hispanic - '+str(r), r1, int(curr+63)]
        finalrace[curr] = ['Hispanic - '+str(r), r2, int(curr)]
        curr+=1
    #sort by percent of pop
    finalrace.sort(key=lambda x:x[1])
    finalrace.reverse()

    
    real_order = []
    for i in range(0, len(finalrace)):
        if i%2 == 0:
            real_order.append(finalrace[i][2])
        if i%2 == 1:
            real_order.insert(0, finalrace[i][2])
            
    #real_order = list(np.arange(0,126,1))
    #print(real_order)
    
    return real_order
```

```python
##########################################################################################################################
#CREATE BLOCK DATA
#input: csv containing block data 
#output: dataframe containing block data for age, sex, race, etc
##########################################################################################################################

def create_block_data(path):
    df = pd.read_csv(path)
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
    return df
```
