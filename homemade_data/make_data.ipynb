```python
import pandas as pd
import csv
import random
```

```python
counties = ['alameda', 'armstrong', 'cibola', 'fayette', 'grandforks', 'hawaii', 'jefferson', 'nantucket', 'washington']
```

```python
###########################################################################################################################
#CREATE DATA - based upon real block & county distributions 
###########################################################################################################################

for county in counties:
    print(county)
    
    #RACE
    #import race data into pdframe
    df = pd.read_csv (r'../2010_census_data/race_data/'+county+'-race.csv')
    #get total of each race description
    race_total = []
    c = 0
    for (columnName, columnData) in df.iteritems():
        if c > 1:
            race_total.append([columnName, sum(columnData.values)])
        c+=1
    
    
    
    #HISPANIC    
    #import race data into pdframe
    dfh = pd.read_csv (r'../2010_census_data/hispanic_data/'+county+'-hispanic.csv').T
    #get total of each race description
    hisp_total = [[],[],[],[],[],[],[], []]
    for row in dfh.iterrows():
            if 'White alone' in row[0]:
                hisp_total[0].append(row[1].iloc[0])
            elif 'Black or African American alone' in row[0]:
                hisp_total[1].append(row[1].iloc[0])
            elif 'American Indian and Alaska Native alone' in row[0]:
                hisp_total[2].append(row[1].iloc[0])
            elif 'Asian alone' in row[0]:
                hisp_total[3].append(row[1].iloc[0])
            elif 'Native Hawaiian and Other Pacific Islander alone' in row[0]:
                hisp_total[4].append(row[1].iloc[0])
            elif 'Some Other Race alone' in row[0]:
                hisp_total[5].append(row[1].iloc[0])
            elif 'Two races including Some other race' in row[0]:
                hisp_total[6].append(row[1].iloc[0])
            elif 'Two races excluding Some other race, and three or more races' in row[0]:
                hisp_total[7].append(row[1].iloc[0])

    #AGE & SEX    
    #import age/sex data into pdframe
    dfas= pd.read_csv (r'../2010_census_data/sexage_data/'+county+'.csv')
    dfas = dfas.dropna()
    pick_sexage = []
    random_offset = [0,1,2,3,4]
    c = 0
    for (columnName, columnData) in dfas.iteritems():
        if c > 1:
            if(columnName[0]=="M"):
                sexage = "m"
            else:
                sexage="f"
            age = int(columnName[2:4])
            add = int(columnData.values * 10)

            for i in range(0,add-1):
                #pick random value 0-4 from array to add 
                new_age = age + random.choice(random_offset)
                new_sexage = sexage + "_" + str(new_age)
                pick_sexage.append(str(new_sexage))
        c+=1   
        
    
    #HOUSEHOLD SIZE
    
    #import household size data into pdframe

    dfhs= pd.read_csv (r'../2010_census_data/householdsize_data/'+county+'-size.csv')
    dfhs = dfhs.dropna()

    pick_size = []

    d = 0
    for (columnName, columnData) in dfhs.iteritems():
        datavals = str(columnData.values).replace(',','').replace('[','').replace(']','').replace('\'','')
        if(' 1' in columnName):
            one_person = int(datavals)
        if(' 2' in columnName):
            two_person = int(datavals)
        if(' 3' in columnName):
            three_person = int(datavals)
        if(' 4' in columnName):
            four_person = int(datavals)


    sum_all = one_person + two_person + three_person + four_person
    per_one = one_person/sum_all * 100
    per_two = two_person/sum_all * 100
    per_three = three_person/sum_all * 100
    per_four = four_person/sum_all * 100

    h = 0
    for p in [per_one, per_two, per_three, per_four]:
        h+=1
        for i in range(0,int((p-1)*10)):
            for i in range(0,h):
                pick_size.append(str(h))
                
                
    #HOUSEHOLD TENURE
    
    #import household tenure data into pdframe 

    dfht= pd.read_csv (r'../2010_census_data/householdtenure_data/'+county+'-tenure.csv')
    dfht = dfht.dropna()

    pick_tenure = []

    t=0
    for (columnName, columnData) in dfht.iteritems():
        datavals = str(columnData.values).replace(',','').replace('[','').replace(']','').replace('\'','')
        if(' Own' in columnName):
            own = int(datavals)
        if(' Rent' in columnName):
            rent = int(datavals)

    sum_all = own+rent
    per_own = own/sum_all * 100
    per_rent = rent/sum_all * 100

    o = 0
    for p in [per_own, per_rent]:
        o+=1
        for i in range(0,int((p-1)*10)):
            if p == per_own:
                pick_tenure.append('Own')
            else:
                pick_tenure.append('Rent')
                
      
    #MAKE DATA
    num = 0
    d = 0
    data= []

    for rval in race_total:
        if d >0:
            times = rval[1]
            if times >=1:
                for i in range(0,int(times)):
                    #pick race
                    r = rval[0]
                    
                    #pick if hispanic or not
                    
                    if 'White alone' in r:
                        hval = hisp_total[0]
                    elif 'Black or African American alone' in r:
                        hval = hisp_total[1]
                    elif 'American Indian and Alaska Native alone' in r:
                        hval = hisp_total[2]
                    elif 'Asian alone' in r:
                        hval = hisp_total[3]
                    elif 'Native Hawaiian and Other Pacific Islander alone' in r:
                        hval = hisp_total[4]
                    elif 'Some other race' in r and 'Two or More Races' in r:
                        hval = hisp_total[6] 
                    elif 'Some Other Race alone' in r:
                        hval = hisp_total[5]
                    elif 'Two or More Races' in r:
                        hval = hisp_total[7]                    
                    
                    hisp = []
                    for i in range(0,int(hval[0]*1000)):
                        hisp.append(0)
                    for i in range(0,int(hval[1]*1000)):
                        hisp.append(1)
                    
                    h = random.choice(hisp)
                            
                    
                    #pick sex and age
                    sa = random.choice(pick_sexage)
                    split = sa.find("_")
                    a = sa[split+1:]
                    s = sa[0:split]
                    
                    

                    #pick 
                    hs = random.choice(pick_size)
                    ht = random.choice(pick_tenure)

                    new_row = [num+1, a, s, r, h, hs, ht]
                    data.append(new_row)

                    num +=1
        d+=1               


        
    #MAKE CSV
    
    col_names = ['id', 'age', 'sex', 'race', 'hispanic', 'household_size', 'household_tenure']
    data = pd.DataFrame(data, columns = col_names)
    
    csv_data = data.to_csv(county+'.csv', index = True)
```
