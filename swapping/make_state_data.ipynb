```python
import pandas as pd
import csv
import random
import numpy as np
```

```python
states = ["california", "georgia", "hawaii", "massachussets", "missouri", "newmexico", "northdakota", "pennsylvania", "vermont"]
```

```python
#MAKE STATE DATA FOR ALL STATES

for state in states:
    
    print(state) 

    data= []
    
    #amount to offset age by
    random_offset = [0,1,2,3,4]

    
#####READ IN DATA##########################################################################################################   
    #read in race data
    df_r = pd.read_csv (r'raw_state_data/'+state+'/race_data.csv')
    #read in sex/age data
    df_as = pd.read_csv (r'raw_state_data/'+state+'/agesex_data.csv')
    #read in household size data
    df_hs = pd.read_csv (r'raw_state_data/'+state+'/householdsize_data.csv')
    #read in household tenure data
    df_ht = pd.read_csv (r'raw_state_data/'+state+'/householdtenure_data.csv')
    #read in hispanic data
    df_h = pd.read_csv (r'raw_state_data/'+state+'/hispanic_data.csv').T

    
#####RACE DATA#############################################################################################################    
   
    #race total will be an array of [race name, real % in state population]
    race_total = []

    total=0
    c = 0
    for (columnName, columnData) in df_r.iteritems():
        if c > 0:
            val = float(str(columnData.values).replace('\'','').replace('[','').replace(']','').replace(',',''))
            total = total+val
        c+=1
    c=0
    for (columnName, columnData) in df_r.iteritems():
        if c>0:
            val = float(str(columnData.values).replace('\'','').replace('[','').replace(']','').replace(',',''))
            race_total.append([columnName.replace('\xa0',''), val/total * 100])
        c+=1

#####HISPANIC & RACE########################################################################################################    

    #hisp total will be an array of race types' hispanic proportions.
    #ex. hisp_total[0] is for white alone; [not hispanic, hispanic] ex. [.1, .9]
    hisp_total = [[],[],[],[],[],[],[], []]
    for row in df_h.iterrows():
            if 'White alone' in row[0]:
                hisp_total[0].append(row[1].iloc[0])
            elif 'Black or African American alone' in row[0]:
                hisp_total[1].append(row[1].iloc[0])
            elif 'American Indian and Alaska Native alone' in row[0]:
                hisp_total[2].append(row[1].iloc[0])
            elif 'Asian alone' in row[0]:
                hisp_total[3].append(row[1].iloc[0])
            elif 'Native Hawaiian and Other Pacific Islander alone' in row[0]:
                hisp_total[4].append(row[1].iloc[0])
            elif 'Some Other Race alone' in row[0]:
                hisp_total[5].append(row[1].iloc[0])
            elif 'Two races including Some other race' in row[0]:
                hisp_total[6].append(row[1].iloc[0])
            elif 'Two races excluding Some other race, and three or more races' in row[0]:
                hisp_total[7].append(row[1].iloc[0])
        
    #finalrace is race/hispanic data of size ~1mil
    finalrace = []
    for rval in race_total:
        r = rval[0]
        for i in range(0,int(rval[1]*10000)):                                       
            if 'White alone' in r:
                hval = hisp_total[0]
            elif 'Black or African American alone' in r:
                hval = hisp_total[1]
            elif 'American Indian and Alaska Native alone' in r:
                hval = hisp_total[2]
            elif 'Asian alone' in r:
                hval = hisp_total[3]
            elif 'Native Hawaiian and Other Pacific Islander alone' in r:
                hval = hisp_total[4]
            elif 'Some other race' in r and 'Two or More Races' in r:
                hval = hisp_total[6] 
            elif 'Some Other Race alone' in r:
                hval = hisp_total[5]
            elif 'Two or More Races' in r:
                hval = hisp_total[7]                    
                    
            hisp = []
            #create 1000 distribution following the hispanic/nonhispanic distribution
            for i in range(0,int(hval[0]*1000)):
                hisp.append(0)
            for i in range(0,int(hval[1]*1000)):
                hisp.append(1)
                    
            #make a random choice from 1000 distribution to determine hispanic/nonhispanic
            h = random.choice(hisp)
            finalrace.append([r,h])
            
        
##### AGE & SEX ########################################################################################################## 
               
    #get total % for age/sex
    agesex_total = []
    total=0
    c=0
    for (columnName, columnData) in df_as.iteritems():
        if c ==1:
            total = float(str(columnData.values).replace('\'','').replace('[','').replace(']','').replace(',',''))
        if c > 1:
            val = float(str(columnData.values).replace('\'','').replace('[','').replace(']','').replace(',',''))
            agesex_total.append([columnName, val/total * 100])
        c+=1

    #finalagesex is 1 mil age/sex data
    finalagesex = []
    for r in agesex_total:
        for i in range(0,int(r[1]*10000)):
            finalagesex.append(r[0])


##### HOUSEHOLD SIZE ###################################################################################################### 

            
    #get total % for each household size
    hs_total = []
    total=0
    for (columnName, columnData) in df_hs.iteritems():
        val = float(str(columnData.values).replace('\'','').replace('[','').replace(']','').replace(',',''))
        total = total+val
    for (columnName, columnData) in df_hs.iteritems():
        val = float(str(columnData.values).replace('\'','').replace('[','').replace(']','').replace(',',''))
        hs_total.append([columnName.replace('\xa0',''), val/total * 100])

    #create household size data of size 1 mil
    finalhs = []
    for r in hs_total:
        for i in range(0,int(r[1]*10000)):
            finalhs.append(r[0])   


##### HOUSEHOLD TENURE ################################################################################################### 

            
    #get total % for each household tenure
    ht_total = []
    total=0
    for (columnName, columnData) in df_ht.iteritems():
        val = float(str(columnData.values).replace('\'','').replace('[','').replace(']','').replace(',',''))
        total = total+val
    for (columnName, columnData) in df_ht.iteritems():
        val = float(str(columnData.values).replace('\'','').replace('[','').replace(']','').replace(',',''))
        ht_total.append([columnName.replace('\xa0',''), val/total * 100])

    #create 1 mil tenure distribution
    finalht = []
    for r in ht_total:
        for i in range(0,int(r[1]*10000)):
        #for i in range(0,int(r[1]*10000)):
            finalht.append(r[0])       

##### COMBINE DATA INTO 1 SET ###################################################################################################### 

    #get min length
    lens = [len(finalrace), len(finalagesex), len(finalhs), len(finalht)]
    minlen = min(lens)


    #combine data into one set
    num=0
    for l in range(0,minlen):
        #choose random of each
        rand_idx = np.random.randint(0, len(finalrace))
        r = finalrace[rand_idx][0]
        h = finalrace[rand_idx][1]
        finalrace.pop(rand_idx)
        rand_idx = np.random.randint(0, len(finalagesex))
        a_s = finalagesex[rand_idx]
        if(a_s[0]=="M"):
            s = "m"
        else:
            s="f"
        age = int(a_s[2:4])
        a = age + random.choice(random_offset)
        finalagesex.pop(rand_idx)
        rand_idx = np.random.randint(0, len(finalhs))
        hs = finalhs[rand_idx]
        finalhs.pop(rand_idx)
        rand_idx = np.random.randint(0, len(finalht))
        ht = finalht[rand_idx]
        finalht.pop(rand_idx)
        new_row = [num+1, a, s, r, h, hs, ht]
        data.append(new_row)
        num +=1
        
    #create csv file
    col_names = ['id', 'age', 'sex', 'race', 'hispanic', 'household_size', 'household_tenure']
    data = pd.DataFrame(data, columns = col_names)
    csv_data = data.to_csv('state_data/'+state+'.csv', index = True)
```
